# Prep Student Performance

```python
import pandas as pd
pd.set_option("display.max_columns",0)
fname = "../Data/student-performance-combined.xlsx"
with  pd.ExcelFile(fname) as f:
    sheet_names = f.sheet_names
    
sheet_names
```

## Version A) Split Into 2 CSV's

```python
import os
OUTPUT_FOLDER = "../Data/Student-Performance-split/"
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
os.listdir(OUTPUT_FOLDER)
```

```python
df_mat = pd.read_excel(fname, sheet_name='student-mat')
df_mat['Subject'] = "Math"
df_mat
```

```python
fname_out = OUTPUT_FOLDER + "student-performance-math.csv"
fname_out
```

```python
df_mat.to_csv(fname_out, index=False)
```

```python
pd.read_csv(fname_out)
```

```python
df_por = pd.read_excel(fname, sheet_name='student-por')
df_por['Subject'] = "Portugese"
df_por
```

```python
fname_out = OUTPUT_FOLDER + "student-performance-por.csv"
fname_out
```

```python
df_por.to_csv(fname_out, index=False)
```

```python
pd.read_csv(fname_out)
```

## Version B) Merged and Melted

```python
df = pd.concat([df_mat, df_por])
df
```

```python
id_cols = df.drop(columns=['G1','G2','G3']).columns
id_cols
```

```python
melted = pd.melt(df, id_vars=id_cols, value_name="Grade",var_name="Period")
melted
```

```python
OUTPUT_FOLDER ='../Data/Student-Performance-melted/'
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
fname_out = OUTPUT_FOLDER + 'student-performance-combined-melted.csv'
fname_out
```

```python
melted.to_csv(fname_out, index=False)
```
